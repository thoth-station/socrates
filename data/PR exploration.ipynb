## Reviewers data

Look at some reviews collected example data

```python
prs_cleared["reviews"][19886]
```

```python
def get_reviews(row):
    data = []
    reviews = row.reviews.values()
    for rev in reviews:
        data.append([row.name, row.created_by, rev["author"], rev["words_count"], rev["submitted_at"], rev["state"]])
    return data
```

```python
not_empty = prs_cleared["reviews"] != {}
```

```python
reviews = prs_cleared[not_empty].apply(lambda x:get_reviews(x),axis=1)
reviews
```

```python
reviews_list = reviews.explode().tolist()
reviews_list
```

```python
pr_reviews = pd.DataFrame.from_records(reviews_list)
pr_reviews.columns = ["pr_id", "pr_author", "review_author", "words_count", "submitted_at", "state"]
```

```python
pr_reviews["date"] = pd.to_datetime(pr_reviews.submitted_at, unit="s")
```

```python
pr_reviews_dates = pr_reviews.set_index("date").drop("submitted_at", axis=1)
```

```python
print(pr_reviews_dates.shape, pr_reviews.shape)
```

```python
pr_reviews_dates
```

### Comment / Approve comparisson for reviewers

```python
approvals = revs_last_6_months["state"] == "APPROVED"
comments = revs_last_6_months["state"] == "COMMENTED"

approvers = revs_last_6_months[approvals].groupby("review_author").agg({"one_vec":"sum"})
commentators = revs_last_6_months[comments].groupby("review_author").agg({"one_vec":"sum"})

approvers.columns = ["approve_count"]
commentators.columns = ["comment_count"]
```

```python
approvers.join(commentators,on="review_author",how="outer").set_index("review_author").sort_values(by="approve_count",ascending=False).plot.bar()
```

### What we can see?
Reviewers tend to comment more and review less (which is expected)
and that means the recommendation metrics can be based on comments alone too
