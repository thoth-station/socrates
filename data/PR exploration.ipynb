## File Ranking recommendation

```python
def get_file_ranks(df):
    file_ranks = {}
    for file,reviewer in zip(df.changed_files, df.review_author):

        if file not in file_ranks:
            file_ranks[file] = {}
        
        if reviewer not in file_ranks[file]:
            file_ranks[file][reviewer] = 1
        else:
            file_ranks[file][reviewer] += 1

        
    for file in file_ranks:
        file_ranks[file] = sorted(list(file_ranks[file].items()), key=lambda x: x[1], reverse=True)
        
    return file_ranks
```

```python
file_ranks = get_file_ranks(exploded_data)
```

```python
def LCP(splitted_path, splitted_other):
    similarity_count = 0

    for first,second in zip(splitted_path, splitted_other):
        if first != second:
            break
        similarity_count += 1
    
    return similarity_count


def get_similarity_score(splitted_path, splitted_other):
    return LCP(splitted_path, splitted_other) / max(len(splitted_path), len(splitted_other))


def get_file_path_similarity_matrix(file_paths):
    similarities = {}
    
    for path in file_paths:
        similarities[path] = {}
        splitted_path = path.split("/")

        for other_path in file_paths:
            if other_path in similarities and path in similarities[other_path]:
                similarities[path][other_path] = similarities[other_path][path]

            splitted_other = other_path.split("/")
            if splitted_path[0] == splitted_other[0]:
                similarities[path][other_path] = get_similarity_score(splitted_path, other_path.split("/"))
            
    return similarities

file_paths = encoders["changed_files"].inverse_transform(X_train.changed_files)
similarity_matrix = get_file_path_similarity_matrix(file_paths)
```

```python
def get_reviewer_scores(file_path, similarity_matrix, file_ranks):
    scores = {}

    for other_path in similarity_matrix[file_path]:

        other_path_id = encoders["changed_files"].transform([file_path])[0]
        for reviewer,score in file_ranks[other_path_id]:
            if reviewer not in scores:
                scores[reviewer] = 0
            scores[reviewer] += similarity_matrix[file_path][other_path] * score
    
    return scores

def recommend_top_x_reviewers(similarity_matrix, file_ranks, file_path, top_x=3):    
    if file_path not in similarity_matrix:
        similarity_matrix[file_path] = {}
        splitted_path = file_path.split("/")
        
        for other_path in similarity_matrix:
            other_splitted = other_path.split("/")
            if splitted_path[0] != other_splitted[0]:
                continue
            similarity_matrix[file_path][other_path] = get_similarity_score(splitted_path, other_splitted)

    reviewer_scores = get_reviewer_scores(file_path, similarity_matrix, file_ranks)
    return sorted(list(reviewer_scores.items()), key=lambda el: el[1])[:top_x]
        
```

### Evaluation

```python
from tqdm import tqdm

def predict(X_set):
    y_prediction = []
    for file in tqdm(encoders["changed_files"].inverse_transform(X_set.changed_files)):
        revs = recommend_top_x_reviewers(similarity_matrix, file_ranks, file)
        y_prediction.append(revs[0])
    return y_prediction

X_train_prediction = predict(X_test)
```
